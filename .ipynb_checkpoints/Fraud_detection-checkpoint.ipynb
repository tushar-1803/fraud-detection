```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_csv("creditcard.csv")

# Display the first few rows of the dataset
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize 'Amount' and 'Time' columns
scaler = StandardScaler()
data['Amount'] = scaler.fit_transform(data[['Amount']])
data['Time'] = scaler.fit_transform(data[['Time']])

# Separate features and labels
X = data.drop(columns=['Class'])
y = data['Class']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train Isolation Forest
iso_forest = IsolationForest(contamination=0.02, random_state=42)
iso_forest.fit(X_train)

# Predict on the test set
y_pred_iso = iso_forest.predict(X_test)

# Convert prediction output to 1 for fraud and 0 for non-fraud
y_pred_iso = [1 if x == -1 else 0 for x in y_pred_iso]

print("Isolation Forest model training complete.")
```

```python
# Evaluate the model
print("Isolation Forest Model Performance:")
print(classification_report(y_test, y_pred_iso, target_names=["Non-Fraud", "Fraud"]))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_iso))
```

```python
# Visualize the class imbalance
import matplotlib.pyplot as plt
import seaborn as sns

# Count plot for fraud vs. non-fraud
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=data)
plt.title("Class Distribution (Fraud vs Non-Fraud)")
plt.xlabel("Class (0 = Non-Fraud, 1 = Fraud)")
plt.ylabel("Count")
plt.show()

# Check for missing values
print("Missing values in each column:\n", data.isnull().sum())

# Correlation matrix to explore feature relationships
plt.figure(figsize=(12, 10))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, fmt=".2f")
plt.title("Correlation Matrix of Features")
plt.show()
```

```python
# Transaction Hour feature
data['TransactionHour'] = (data['Time'] // 3600) % 24  # Convert seconds to hours (mod 24)

# Amount Binning (low, medium, high)
data['AmountBin'] = pd.cut(data['Amount'], bins=[-1, 50, 200, np.inf], labels=['Low', 'Medium', 'High'])

# Log transformation of Amount
data['LogAmount'] = np.log1p(data['Amount'])  # log1p to handle zero values

# Drop the original 'Time' column if we're using the new 'TransactionHour' feature
data = data.drop(columns=['Time'])

# Check the first few rows to see the new features
data.head()
```

```python
# Perform one-hot encoding on 'AmountBin'
data = pd.get_dummies(data, columns=['AmountBin'], drop_first=True)

# Separate features and target variable again
X = data.drop(columns=['Class'])
y = data['Class']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Apply SMOTE to balance the classes in the training set
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Check the class distribution after SMOTE
print("Class distribution after SMOTE:")
print(y_train_smote.value_counts())
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the autoencoder model
input_dim = X_train_smote.shape[1]

autoencoder = Sequential([
    Dense(32, activation='relu', input_shape=(input_dim,)),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(16, activation='relu'),
    Dense(32, activation='relu'),
    Dense(input_dim, activation='sigmoid')
])

autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder on non-fraudulent data only
X_train_non_fraud = X_train_smote[y_train_smote == 0]
history = autoencoder.fit(X_train_non_fraud, X_train_non_fraud, epochs=20, batch_size=32, validation_split=0.1, verbose=1)
```

```python
# Calculate reconstruction errors on the training data (non-fraudulent)
X_train_pred = autoencoder.predict(X_train_non_fraud)
reconstruction_error = np.mean(np.power(X_train_non_fraud - X_train_pred, 2), axis=1)

# Set the threshold at the 95th percentile
threshold = np.percentile(reconstruction_error, 95)
print("Reconstruction error threshold:", threshold)
```

```python
# Calculate reconstruction errors for the test set
X_test_pred = autoencoder.predict(X_test)
test_error = np.mean(np.power(X_test - X_test_pred, 2), axis=1)

# Classify as fraud if the error exceeds the threshold
y_pred_autoencoder = [1 if error > threshold else 0 for error in test_error]

# Evaluate the model's performance
print("Autoencoder Model Performance:")
print(classification_report(y_test, y_pred_autoencoder, target_names=["Non-Fraud", "Fraud"]))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_autoencoder))
```

```python
# Set the threshold at the 90th percentile to make the model more sensitive
new_threshold = np.percentile(reconstruction_error, 90)
print("New reconstruction error threshold:", new_threshold)

# Classify as fraud if the error exceeds the new threshold
y_pred_autoencoder_new = [1 if error > new_threshold else 0 for error in test_error]

# Evaluate the model's performance with the new threshold
print("Autoencoder Model Performance with Adjusted Threshold:")
print(classification_report(y_test, y_pred_autoencoder_new, target_names=["Non-Fraud", "Fraud"]))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_autoencoder_new))
```

```python
from tensorflow.keras.layers import Dropout

# Define an enhanced autoencoder model with additional layers and dropout
enhanced_autoencoder = Sequential([
    Dense(64, activation='relu', input_shape=(input_dim,)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(16, activation='relu'),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dense(input_dim, activation='sigmoid')
])

enhanced_autoencoder.compile(optimizer='adam', loss='mse')

# Train the enhanced autoencoder on non-fraudulent data
history = enhanced_autoencoder.fit(X_train_non_fraud, X_train_non_fraud, epochs=20, batch_size=32, validation_split=0.1, verbose=1)
```

```python
# Calculate reconstruction errors on the non-fraudulent training data
X_train_pred_enhanced = enhanced_autoencoder.predict(X_train_non_fraud)
reconstruction_error_enhanced = np.mean(np.power(X_train_non_fraud - X_train_pred_enhanced, 2), axis=1)

# Set the new threshold at the 90th percentile
new_threshold_enhanced = np.percentile(reconstruction_error_enhanced, 90)
print("Enhanced model reconstruction error threshold:", new_threshold_enhanced)
```

```python
# Calculate reconstruction errors for the test set
X_test_pred_enhanced = enhanced_autoencoder.predict(X_test)
test_error_enhanced = np.mean(np.power(X_test - X_test_pred_enhanced, 2), axis=1)

# Classify as fraud if the error exceeds the new threshold
y_pred_autoencoder_enhanced = [1 if error > new_threshold_enhanced else 0 for error in test_error_enhanced]

# Evaluate the model's performance with the enhanced architecture
print("Enhanced Autoencoder Model Performance:")
print(classification_report(y_test, y_pred_autoencoder_enhanced, target_names=["Non-Fraud", "Fraud"]))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_autoencoder_enhanced))
```

```python
# Fit Isolation Forest on the training data
iso_forest = IsolationForest(contamination=0.02, random_state=42)
iso_forest.fit(X_train)

# Predict on the test set
y_pred_iso_forest = iso_forest.predict(X_test)
y_pred_iso_forest = [1 if pred == -1 else 0 for pred in y_pred_iso_forest]  # Convert to binary fraud labels
```

```python
# Ensemble prediction (intersection of autoencoder and isolation forest predictions)
y_pred_ensemble = [1 if (autoencoder_pred == 1 and iso_pred == 1) else 0 for autoencoder_pred, iso_pred in zip(y_pred_autoencoder_enhanced, y_pred_iso_forest)]

# Evaluate the ensemble model's performance
print("Ensemble Model Performance (Intersection of Autoencoder and Isolation Forest):")
print(classification_report(y_test, y_pred_ensemble, target_names=["Non-Fraud", "Fraud"]))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_ensemble))
```

```python
# Ensemble prediction (union of autoencoder and isolation forest predictions)
y_pred_union = [1 if (autoencoder_pred == 1 or iso_pred == 1) else 0 for autoencoder_pred, iso_pred in zip(y_pred_autoencoder_enhanced, y_pred_iso_forest)]

# Evaluate the ensemble model's performance with the union approach
print("Ensemble Model Performance (Union of Autoencoder and Isolation Forest):")
print(classification_report(y_test, y_pred_union, target_names=["Non-Fraud", "Fraud"]))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_union))
```

```python
# Define weights for each model
autoencoder_weight = 0.7
iso_forest_weight = 0.3
combined_threshold = 0.5  # Adjust this threshold as needed

# Calculate the weighted score
weighted_scores = [
    (autoencoder_weight * autoencoder_pred) + (iso_forest_weight * iso_pred)
    for autoencoder_pred, iso_pred in zip(y_pred_autoencoder_enhanced, y_pred_iso_forest)
]

# Classify as fraud if the combined score exceeds the threshold
y_pred_weighted = [1 if score >= combined_threshold else 0 for score in weighted_scores]

# Evaluate the model's performance with the weighted scoring approach
print("Ensemble Model Performance (Weighted Scoring Approach):")
print(classification_report(y_test, y_pred_weighted, target_names=["Non-Fraud", "Fraud"]))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_weighted))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train Random Forest on SMOTE-balanced training data
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_smote, y_train_smote)

# Predict on the test set with Random Forest
y_pred_rf = rf_model.predict(X_test)
```

```python
# Convert Autoencoder, Isolation Forest, and Random Forest predictions to a majority voting ensemble
y_pred_ensemble_majority = [
    1 if (autoencoder_pred + iso_pred + rf_pred) >= 2 else 0
    for autoencoder_pred, iso_pred, rf_pred in zip(y_pred_autoencoder_enhanced, y_pred_iso_forest, y_pred_rf)
]

# Evaluate the ensemble model's performance with majority voting
print("Ensemble Model Performance (Majority Voting of Autoencoder, Isolation Forest, and Random Forest):")
print(classification_report(y_test, y_pred_ensemble_majority, target_names=["Non-Fraud", "Fraud"]))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_ensemble_majority))
```
